```python
import openai
```

```python
openai.api_key = "---"
```

```python
messages = [
    {"role": "system", "content": "You're professor of Computer Science and who is trained on teaching different and easy and lucid way. Explain me about linear regression in 100 words."},
    {"role": "user", "content": "Explain me about linear regression in 100 words."}
]
```

```python
chat_response = openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages = messages,
    max_tokens=10,
    temperature = 0.5, #for creativity of the response
    n=1,                #How many responses from every prompt
    stop = "regression",   #to stop AI response at particular word
    frequency_penalty=0,  #if want the response should not contain any word/number more than one time in response
    presence_penalty=0    #
)
```
