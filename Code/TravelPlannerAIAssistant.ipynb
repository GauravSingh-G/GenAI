```python
message_history = [
    {
        "role": "system",
        "content": "You are a travel itinerary assistant. Create personalized plans.\nRequire: budget, destination, travelers.\nIf missing, ask only for missing info; do not generate yet.\nOnce complete, generate itinerary.\nInclude: day-wise plan, stays, budget, tips.\nTone: friendly, helpful, professional."
    },
    {"role": "system", "name": "example_user", "content": "I want to go on a trip"},
    {"role": "system", "name": "example_assistant", "content": "Nice! Where do you want to go?"},

    {"role": "system", "name": "example_user", "content": "Agra"},
    {"role": "system", "name": "example_assistant", "content": "Great! What's your budget and number of travelers?"},

    {"role": "system", "name": "example_user", "content": "8000 for 2 people"},
    {"role": "system", "name": "example_assistant", "content": "Got it! When are you planning to travel?"},

    {"role": "system", "name": "example_user", "content": "Next weekend"},
    {"role": "system", "name": "example_assistant", "content": "Here's your itinerary:\nDestination: Agra\nBudget: Rs.8000\nPeople: 2\nDates: Next weekend\nDay 1: Arrival, Agra Fort\nDay 2: Taj Mahal, market, depart\nTips: Book early, rent scooter, keep phone safe."},

    {"role": "user", "content": "I want to plan a trip to Manali"}
]
```

```python
max_conversations = 10
conversation_length = 0
```

```python
conversation_length = 0
max_conversations = 10

while conversation_length < max_conversations:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("AI: Exiting the program!")
        break

    message_history.append({"role": "user", "content": user_input})

    response = openai.chat.completions.create(
        model="gpt-3.5-turbo-16k",
        messages=message_history
    ).choices[0].message.content

    print("\nAI:")
    print(response)
    print()

    message_history.append({"role": "assistant", "content": response})

    conversation_length += 1
```
