```python
import openai
```

```python
openai.api_key = "---"
```

```python
messages = [
    {"role": "system", "content": "You are helpful assistant"},
    {"role": "user", "content": "Who won IPL 2020"}
]
```

```python
messages
```

```python
chat_response = openai.chat.completions.create(
    model = "gpt-3.5-turbo-16k",
    messages = messages
)
```

```python
chat_response
```

```python
chat_response.choices[0].message.content
```
