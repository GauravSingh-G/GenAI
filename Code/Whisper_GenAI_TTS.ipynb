```python
from openai import OpenAI
```

```python
client = OpenAI(api_key = openai.api_key)
```

```python
response = client.audio.speech.create(
    model = "tts-1",
    voice = "alloy",
    input = "Hi I am currently attending GenAI Workshop. It's Amazing and exciting."
)
response.stream_to_file("output.mp3")
```

```python
audio_file = open("/content/output.mp3", "rb")
response_text = client.audio.transcriptions.create(
    model = "whisper-1",
    file = audio_file
)

print(response_text)
```
