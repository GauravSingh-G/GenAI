```python
file_path ="/content/drive/MyDrive/GenAI/HCST.txt"
```

```python
with open(file_path, 'r') as file:
    content = file.read()
```

```python
base_instruction='''You are a helpful professional assistant which helps in analyse and retrieve the relevant information
and document information from the documents I provide as input in concise and bestest way.
Given below are some questions and the transcripts of Hindustan college NIRF document,
Try to respond with specific numbers and facts wherever it is possible.
If you are unsure about any question just say 'I don't know dont give any random information.'.'''
```

```python
question=[
    "Q1: The doc shared is of which year?",
    "Q2: No. of students enrolled in the session 2023–2024. Give the detailed breakdown.",
    "Q3: Is there a % increase in number of students enrolled in consecutive years or not? Give the complete comprehensive calculation.",
    "Q4: Give me the complete faculty details who has the designation as 'Professor'.",
    "Q5: No. of total branch students vs PhD holders across all years.",
    "Q6: Give me the detailed breakdown of fee details for B.Tech."
]
```

```python
formatted_questions="\n".join(question)
prompt=base_instruction+"\n\n"+"question:\n{0}".format(formatted_questions)+"\n\n"+"Transcript:\n{0}".format(content)
```

```python
message=[{"role":"user","content":prompt}]
```

```python
chat_response=openai.chat.completions.create(
    model="gpt-5-nano",
    messages=message

)
print(chat_response.choices[0].message.content)
```
