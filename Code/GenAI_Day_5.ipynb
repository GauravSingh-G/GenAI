```python
openai.api_key = "---"
```

```python
client = OpenAI(api_key = openai.api_key)
```

```python
response = client.images.generate(
    model = "dall-e-2",
    prompt = "An apple floating in water",
    n=1,
    size = "1024x1024"

)
image_url = response.data[0].url
print(f"your image is ready at: {image_url}")
```
